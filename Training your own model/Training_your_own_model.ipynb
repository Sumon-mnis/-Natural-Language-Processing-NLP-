```python
import os
import numpy as np
import pandas as pd
import gensim

import nltk
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
story = []

# Get the list of all files in the current directory, assuming they are the ones uploaded
# Alternatively, you could use `uploaded.keys()` if you only want to process the files uploaded via `files.upload()`
file_names = [f for f in os.listdir('/content/') if f.endswith('.txt') and f.startswith('00')]

for filename in file_names:
    file_path = os.path.join('/content/', filename)

    with open(file_path, 'r', encoding='latin-1') as f:
        corpus = f.read()

    sentences = sent_tokenize(corpus)

    for sent in sentences:
        processed_sent = simple_preprocess(sent)
        story.append(processed_sent)
```

```python
len(story)
```

```python
story[:5]
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences=story,
    vector_size=100,
    window=5,
    min_count=2,
    workers=4
)
```

```python
model.wv.most_similar("king")
```

```python
model.save("word2vec_model.model")
```
